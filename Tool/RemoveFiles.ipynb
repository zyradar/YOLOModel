清除无标签图片

```python
import os

def clear(imgDir):
    names = os.listdir(imgDir)
    for name in names:
        imgPath=os.path.join(imgDir,name)
        # find txt
        txtPath = imgPath.replace('.jpg','.txt')
        if not os.path.isfile(txtPath):
            print(imgPath)
            os.unlink(imgPath)
            
imgDir=r'C:\Users\LanyeSaint\Desktop\Code\Python\ultralytics-main\datasets'
clear(imgDir)
```
